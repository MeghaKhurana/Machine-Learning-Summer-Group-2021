## INTRODUCTION

Data Visualization is the third and final topic of Data Analysis that we'll be covering. Data visualization is the graphical representation of information and data. It refers to techniques used to communicate insights from data through visual representation. It's always easier to look at a line plot to understand a trend, than looking at rows and rows of data. Making informative visualizations (sometimes called plots) is one of the most important tasks in data analysis. 

Every good visual must have the following four qualities:
<ul>
    <li><b>Truthfulness</b>: Your plots must always accurately show what the data actually says. Using techniques like changing axis scale to suit your needs, and showing something misleading like only the convenient part of the data in the plot is wrong.</li>
    <li><b>Beauty</b>: Your plots must be appealing to the eye. Beautiful plots catch the audience attention and convey what you are trying to say.</li>
    <li><b>Functionality</b>: You should plot the right plot for the right kind of data so that the functionality of your plot is maximum.</li>
    <li><b>Insightfulmess</b>: This is probably the most important quality a visual must have, at least for an ML engineer. Your visual must be insightful and must reveal new information, trends, patterns, anomalies et. which lie hidden in a table of data.</li>

You might be wondering how this visualization is relevant for ML. That's where EDA steps in. We'll discuss about EDA at the end of this notebook.

### Figures and Axes<br>
<b>Figure</b>: The entire figure which may contain one or more Axes is called the Figure object.<br>
<b>Axes</b>: The Axes object is used to distinctly identify a plot. It may contain 2 (in 2D) or 3 (in 3D) Axis objects.<br>
<b>Axis</b>: The Axis object refers to numer-line like objects.<br><br>
A figure may contain a single plot or many subplots. The figure object is used to access the entire figure. In case the figure has many subplots, then each subplot can be accessed using the Axes object (Note the difference between <b>Axes</b> and <b>Axis</b>).<br>
So if you have a figure which has many subplots, you would use the Axes object to set titles to each subplot separately and use the Figure object to set a title to the entire figure.

### <code>plt.plot()</code>

<code>plt.plot()</code> plots y values against x values in the form of a line plot. (x and y must have same dimensions)

```python
x = np.arange(1,11)
y = x**2
plt.plot(x, y)
```

If you provide a single array as argument, then they are considered as y values and plotted against their indices

```python
#plt.figure() is used to create a new empty figure so that the new plots don't go back to your old plot.
#plt always refers to the last plotted figure
plt.figure()
y = np.arange(1,11)
plt.plot(y)
```

You can also plot multiple lines with a single plot command with different formatting as shown.

```python
plt.figure()

x1 = np.linspace(0,10,100)
x2 = x1
y1 = np.sin(x1)
y2 = np.cos(x2)

plt.plot(x1, y1, "x--k", x2, y2, "o-r")
```

### Figures and subplots

As we have already seen, <code>plt.figure()</code> is used to create a new empty figure.

```python
fig = plt.figure() #Creates an empty figure
```

```python
plt.plot([1,2,3,4], [5,10,5,10])
```

Multiple subplots can be created in a single figure using <code>plt.subplots()</code>

```python
fig, axes = plt.subplots(2,2) #2 rows and 2 columns

#Here fig can be used to access the entire figure. axes is a 2x2 numpy array storing the axes objects for each subplot.
#axes[i][j] makes changes to the subplot in ith row and jth column (0 indexed)

axes[0,0].plot([1,2,3,4], [5,10,5,10])
axes[0,1].plot([1,2,3,4], [1,2,3,4])
axes[1,0].plot(np.random.randn(50).cumsum(), 'k--')
axes[1,1].plot(np.linspace(-1,1,21), abs(np.linspace(-1,1,21)))
```

### <code>plt.scatter()</code>
This command creates a scatter plot with markers of varying size and figure.

```python
plt.figure()

#This function necessarily requires both x and y values
x = np.linspace(-10,11,30)
y = x**2 + np.random.randint(-10,10,30)
plt.scatter(x, y)
```

The size of the markers can be changed based on a third list and in a similar way color of the scatter points can also be changed.

```python
plt.figure()

#This function necessarily requires both x and y values
x = np.linspace(-10,11,30)
y = x**2 + np.random.randint(-10,10,30)
size = np.random.randint(1,50,30)
plt.scatter(x, y, s = size)
```

Checkout the <a href = "https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html">link</a> for more information on various kinds of formatting options in scatter plots.

### Customizing a plot

Most of the times, just plotting values may not be enough. You need to make the plot presentable and understandable by adding title, labels, legends etc.

```python
plt.figure()

#Creating x and y values
x = np.linspace(1,11,50)
y_best = ((2*x) + 4)
y = y_best + np.random.randint(-5,5,50)

#Creating a scatter plot of x and y values with 0.7 opacity
sca = plt.scatter(x,y, alpha = 0.7, c = "#558cf2")

#Creating a line plot of the best fit line
lin,  = plt.plot(x, y_best, "--r", linewidth = 2)
#plt.plot always returns a tuple of all the lines it plots (even if only 1 line is plotted)
#Hence the comma is added after lin to get the line2D object and not the tuple

#Adding x and y label
plt.xlabel("X Values")
plt.ylabel("Y Values")

#Adding a title
plt.title("Simple Best Fit Plot")

#Adding a legend
plt.legend([sca, lin], ["Scattered points", "Best fit line"])

#Adding grid with 0.5 opacity
plt.grid(alpha = 0.9)
```

This clearly looks better and presentable than the crude plots we were drawing previously. Each function here has many customizations of it's own and hence gives you an oppurtunity to unleash your creativity in the plots. There are various other commands to change ticks, ticklabels etc. Most people just find the necessary customization using a single Google Search hence there's no need to memorize any of these functions

### Bar Graphs
A <b>bar graph</b> breaks categorical data down by group, and represents these amounts by using bars of different lengths. It uses either the number of individuals in each group (also called the frequency) or the percentage in each group (called the relative frequency).

```python
#Loading the dataset
df = sns.load_dataset('tips')
df
```

```python
#Plotting bar graph of total tips by day using pandas
plt.figure()

ser = df["tip"].groupby(df["day"]).sum()
ser.plot(kind = 'bar')

plt.title("Total tips by day")
plt.xlabel("Day")
plt.ylabel("Total tips");
```

```python
#Plotting total bill by day using seaborn, it groups the y values by x values

plt.figure()
sns.barplot(x = "day", y = 'total_bill', data = df, capsize = 0.1, estimator = sum)

plt.title("Total bill by day")

#The seaborn plot also shows the 95% confidence interval by default
```

```python
#Plots by nested grouping

plt.figure()
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = df, estimator = np.mean)
```

### Histograms and Density plots

These plots are mainly used for univariate analysis.<br>
A <b>histogram</b> is a kind of bar plot that gives a discretized display of value frequency. The data points are split into discrete, evenly spaced bins, and the number of data points in each bin is plotted.

```python
#For histograms we shall plot normal distributions
x = np.random.randn(1000)
```

```python
#Plotting a histogram using matplotlib
plt.figure()
plt.hist(x, bins = 59, edgecolor = 'black');
#Splits the data into 50 bins
```

```python
#Plotting histogram using seaborn
plt.figure()
sns.distplot(x, kde = False, hist_kws = dict(edgecolor = 'k'));
```

A <b>density plot</b>, is formed by computing an estimate of a continuous probability distribution that might have generated the observed data.

```python
#seaborn distplot by default fits a gaussian kernel density estimate to histogram
plt.figure()
sns.distplot(x);
```

### Heatmaps

A heatmap is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions. A heatmap essentially represents 3D data on 2D space. It is one of the most widely used plots in data analysis.

```python
#Loading the dataset
flight = sns.load_dataset('flights')
flight
```

```python
#Creating a pivot table. Checkout df.pivot()
flight = flight.pivot("month", "year", "passengers")
flight
```

```python
#Plotting the heatmap
plt.figure()
sns.heatmap(flight)
```

From the heatmap we can instanly notice the increase in passengers as years pass by. Aslso a general increase around the middle of the year.
